```python
import numpy as np
import pandas as pd
from plotnine import ggplot, aes, geom_point, geom_bar, geom_line, labs
```

```python
import plotly.graph_objects as go
```

```python
import plotly.express as px
```

 Exercise 1 : Basic Scatter Plot With Plotnine

```python
# Create a pandas DataFrame with two columns, x and y, each containing 20 random numbers.
df = pd.DataFrame({
    'x': np.random.rand(20),
    'y': np.random.rand(20)
})
```

```python
# Plot using Plotnine
scatter_plot = (ggplot(df, aes(x='x', y='y')) +
                geom_point() +
                labs(x='X-Value', y='Y-Value', title='Basic Scatter Plot'))

# Display the scatter plot
print(scatter_plot)
```

Exercise 2 : Bar Plot Of Categorical Data With Plotnine

```python
# Generate a pandas DataFrame with a column category, containing 30 random entries from a set of categories.
df = pd.DataFrame({'category': np.random.choice(['A', 'B', 'C'], size=30)})
# Plot using Plotnine
bar_plot = (ggplot(df, aes(x='category')) +
            geom_bar() +
            labs(x='Category', y='Frequency', title='Category Frequency'))

# Display the bar plot
print(bar_plot)
```

Exercise 3 : Line Plot With Multiple Lines With Plotnine

```python
x = np.linspace(0, 10, 100)
# Calculate y1 and y2 values (e.g., sin(x) and cos(x))
y1 = np.sin(x)
y2 = np.cos(x)
# Create pandas DataFrame with three columns: x, y1, and y2
df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})
# Create a line plot with two lines, one for y1 and one for y2.
line_plot = (
    ggplot(df, aes(x='x', y='y1'))
    + geom_line(aes(y='y1'), color='blue')
    + geom_line(aes(x='x', y='y2'), color='orange')
    + labs(x='X-Value', y='Y-Value', title='Line Plot with Multiple Lines')
)

# Display the line plot
print(line_plot)
```

Exercise 4 : Basic Line Chart With Plotly

```python
# Define monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]

# Create a line chart
fig = go.Figure(data=go.Scatter(x=months, y=sales, mode='lines'))

# Add title
fig.update_layout(title='Monthly Sales Data')

# Display the plot
fig.show()
```

Exercise 5 : Bar Chart With Customization With Plotly

```python
# Define genre data
genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]

# Define a list of colors for the bars.
colors = ['#1F77B4', '#FF7F0E', '#2CA02C', '#D62728', '#9467BD']
# Create a bar chart
fig = go.Figure(data=go.Bar(x=genres, y=books_sold, marker_color=colors))

# Add title
fig.update_layout(title='Books Sold by Genre')

# Display the plot
fig.show()
```

Exercise 6 : Scatter Plot With Size And Color Dimensions With Plotly

```python
# Define the dataset
data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})

# Create a scatter plot
fig = px.scatter(data, x='Age', y='Salary', size='Years of Experience',  color="Years of Experience", title='Relationship between Age, Salary, and Experience')

# Display the plot
fig.show()
```
