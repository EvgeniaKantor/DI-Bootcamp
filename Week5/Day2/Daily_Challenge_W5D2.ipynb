```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("imdb_top_1000.csv")

# Display the first few rows of the DataFrame to understand its structure
print(df.head())
```

```python
# Display the summary information about the DataFrame
print(df.info())
```

```python
# Display basic statistics of the numerical columns
print(df.describe())
```

```python
# Check for any missing values
print(df.isnull().sum())
```

```python
# Get unique values in the Gross column
unique_gross_values = df['Gross'].unique()
print("\nUnique values in Gross column:")
print(unique_gross_values)
```

```python
# Remove commas from the values in the Gross column and convert to numeric type
df['Gross'] = df['Gross'].str.replace(',', '').astype(float)

# Check the data type of the Gross column after conversion
print(df['Gross'].dtype)
```

```python
# Display the first few rows of the DataFrame to understand its structure
print(df.head())
```

```python
# Check for any missing values
print(df.isnull().sum())
```

```python
# Get unique values in the Certificate column
unique_certificates = df['Certificate'].unique()
print("Unique values in Certificate column:")
print(unique_certificates)
```

```python
# Get unique values in the Meta_score column
unique_meta_scores = df['Meta_score'].unique()
print("\nUnique values in Meta_score column:")
print(unique_meta_scores)
```

```python
# Drop rows with missing values in the Gross column
df.dropna(subset=['Gross'], inplace=True)
# Check for any missing values
print(df.isnull().sum())
```

```python
# Get unique values in the Meta_score column
unique_meta_scores = df['Released_Year'].unique()
print("\nUnique values in Released_Year column:")
print(unique_meta_scores)
```

```python
# Drop rows with "PG" in the "Released_Year" column
df = df[df['Released_Year'] != 'PG']

# Convert the "Released_Year" column to datetime with format specifier %Y
df['Released_Year'] = pd.to_datetime(df['Released_Year'], format='%Y')

# Check the unique values in the "Released_Year" column again
unique_released_years = df['Released_Year'].dt.year.unique()
print("\nUnique values in Released_Year column:")
print(unique_released_years)
```

```python
# Extract only the "Meta_score" and "IMDB_Rating" columns into a new DataFrame
ratings_df = df[['Meta_score', 'IMDB_Rating']].copy()

# Drop rows with missing values in either Meta_score or IMDB_Rating
ratings_df.dropna(inplace=True)

# Calculate the correlation coefficient
correlation = ratings_df['Meta_score'].corr(ratings_df['IMDB_Rating'])

print("Correlation between Meta_score and IMDB_Rating:", correlation)
```


The correlation coefficient between "Meta_score" and "IMDB_Rating" is approximately 0.283. This positive correlation indicates that there is a weak positive linear relationship between the IMDb rating of a movie and its Metascore. However, the correlation is not very strong, suggesting that while there may be some relationship between these two variables, it is not particularly strong. So we don't have enough information to fill in the missing Meta_score values

```python
# Drop rows with missing values in the Meta_score column
df.dropna(subset=['Meta_score'], inplace=True)
# Check for any missing values
print(df.isnull().sum())
```

```python
# Display the summary information about the DataFrame
print(df.info())
```

```python
df['Certificate'] = df['Certificate'].fillna('')
df.isna().sum()
```

```python
# Check the unique values in the "Runtime" column again
unique_runtime = df['Runtime'].unique()
print("Unique values in Runtime column:")
print(unique_runtime)
```

```python
# Remove "min" from each value in the "Runtime" column
df['Runtime'] = df['Runtime'].str.replace(' min', '')

# Convert the "Runtime" column to integer type
df['Runtime'] = df['Runtime'].astype(int)

# Check the unique values in the "Runtime" column again
unique_runtimes = df['Runtime'].unique()
print("\nUnique values in Runtime column after conversion:")
print(unique_runtimes)
```

```python
# Group the data by 'Released_Year' and calculate the mean IMDB_Rating for each year
rating_trends = df.groupby('Released_Year')['IMDB_Rating'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(rating_trends.index, rating_trends.values, marker='o', linestyle='-')
plt.title('Trends in IMDB Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean IMDB Rating')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the number of movies in each genre
genres = df['Genre'].str.split(', ', expand = True).stack()

# Count the occurrences of each genre
genre_counts = genres.value_counts()
# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, hue=genre_counts, palette='muted')
plt.title('Number of Movies Across Different Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
# Group data by director and calculate average IMDB_Rating
average_ratings_by_director = df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False)

# Select top 15 directors with the highest average ratings
top_directors = average_ratings_by_director.head(15)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(top_directors.index, top_directors.values, color='blue')
plt.xlabel('Directors')
plt.ylabel('Average IMDB Rating')
plt.title('Relationship between Directors and Movie Ratings')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Select top 15 directors with the highest ratings
top_directors = average_ratings_by_director.head(15)

# Create a scatter plot
plt.figure(figsize=(12, 8))
for director in top_directors.index:
    director_movies = df[df['Director'] == director]
    plt.scatter(director_movies['Director'], director_movies['IMDB_Rating'], label=director, alpha=0.7)

plt.xlabel('Director')
plt.ylabel('IMDB Rating')
plt.title('IMDB Rating by Director (Top 15 Directors)')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.tight_layout()
plt.show()
```

```python

top_stars = df['Star1'].value_counts().head(5).index

stars_and_ratings = df[df['Star1'].isin(top_stars)]
stars_and_ratings = stars_and_ratings[['IMDB_Rating', 'Gross', 'Star1', 'Star2', 'Star3', 'Star4']]
sns.pairplot(stars_and_ratings, height=6, aspect=0.7, hue='Star1')
plt.suptitle('Relationships between Stars, IMDB Rating, and Gross')
plt.show()
```

```python
top_stars = df['Star2'].value_counts().head(5).index

stars_and_ratings = df[df['Star2'].isin(top_stars)]
stars_and_ratings = stars_and_ratings[['IMDB_Rating', 'Gross', 'Star1', 'Star2', 'Star3', 'Star4']]
sns.pairplot(stars_and_ratings, height=6, aspect=0.7, hue='Star2')
plt.suptitle('Relationships between Stars, IMDB Rating, and Gross')
plt.show()
```

```python
# Split the genres column into separate entries
genres = df['Genre'].str.split(', ', expand=True).stack()

# Reset index to align with genres
df_genre = df.loc[genres.index.get_level_values(0)].copy()

# Assign the genres to the DataFrame
df_genre['Genre'] = genres.values

# Create the box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Genre', y='IMDB_Rating', data=df_genre, palette='Set3')
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of IMDb Ratings Across Different Genres')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating')
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_columns = ['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']

# Calculate correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

high cor for 'IMDB_Rating' and 'No_of_Votes' = 0.61 (between Rating of the movie at IMDB site and Total number of votes)
and for 'Gross' and 'No_of_Votes' = 0.56 (between  Money earned by that movie and Total number of votes)
