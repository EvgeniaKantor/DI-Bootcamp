**Exercise 1 : Understanding Data Visualization**

**Why data visualization is important:**


**Reveals Insights:** Visual representations help uncover patterns, trends, and anomalies in data quickly.

**Facilitates Communication:** Visualizations make it easier to present complex findings to stakeholders in an understandable way.

**Aids Decision Making:** By presenting data visually, decision-makers can grasp implications more effectively.


**Purpose of a line graph:**


**Display Trends:** They show how variables change over time or across a continuous interval.

**Enable Comparison:** Line graphs allow comparison between multiple groups or categories.

**Highlight Changes:** They effectively highlight shifts or anomalies in data trends.

**Facilitate Forecasting:** Line graphs aid in predicting future trends based on historical data.

```python
# Exercise 2
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'Temperatures (°F)': [72, 74, 76, 80, 82, 78, 75]
}
df = pd.DataFrame(data)
df
```

```python
# Create a line plot
plt.plot(df['Days'], df['Temperatures (°F)'], marker='o', linestyle='-')

# Add labels and title
plt.ylabel('Temperature (°F)')
plt.title('Temperature Variations Over a Week')

# Show the plot
plt.grid(True)
plt.show()
```

Exercise 3 : Visualizing Monthly Sales With A Bar Chart

```python
data = {
    'Months': ['January', 'February', 'March', 'April', 'May'],
    'Monthly_sales' : [5000, 5500, 6200, 7000, 7500]
}
df = pd.DataFrame(data)
df
```

```python
# Create a bar chart
plt.bar(df['Months'], df['Monthly_sales'], color='skyblue')

# Add labels and title
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales')

# Show the plot
plt.grid(axis='y')
plt.show()
```

```python
#  Exercise 4 : Data Visualisation
import pandas as pd

# URL of the CSV file
url = 'https://raw.githubusercontent.com/intern2grow/sales-data-analysis/main/sales_data.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(url)

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())
```

```python
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Convert 'date' column to datetime data type
df['date'] = pd.to_datetime(df['date'])

# Ensure consistency and correctness in 'product' and 'category' columns
df['product'] = df['product'].str.lower()  # Capitalize product names
df['category'] = df['category'].str.lower()  # Convert category names to lowercase

# Check for outliers in numerical columns
print("\nSummary statistics:")
print(df.describe())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df.head())
```

```python
# Calculate the total quantity of products sold
total_quantity_sold = df['quantity'].sum()
print("Total quantity of products sold:", total_quantity_sold)
```

```python
# Identify the category with the highest revenue
category_highest_revenue = df.groupby('category')['revenue'].sum().idxmax()
highest_revenue = df.groupby('category')['revenue'].sum().max()
print("Category with the highest revenue:", category_highest_revenue)
print("Revenue generated by the highest revenue category:", highest_revenue)
```

```python
# Determine the average revenue per sale
average_revenue_per_sale = df['revenue'].mean()
print("Average revenue per sale:", average_revenue_per_sale)
```

```python
# Extract the quarter from the 'date' column
df['quarter'] = df['date'].dt.quarter

# Calculate total revenue generated in each quarter
revenue_per_quarter = df.groupby('quarter')['revenue'].sum()

# Create a bar chart to visualize total revenue per quarter
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
plt.bar(quarters, revenue_per_quarter, color='skyblue')
plt.title('Total Revenue Generated in Each Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Revenue ($)')
plt.show()
```

```python
#Exercise 5 : Data Visualisation Using MatPlotLib
import matplotlib.pyplot as plt
import numpy as np

# Task 1: Create a line plot for y = x^2
x_values = np.linspace(-10, 10, 100)  # Generate x values ranging from -10 to 10
y_values = x_values ** 2  # Calculate y values
plt.plot(x_values, y_values, color='blue')  # Plot the line
plt.title('Plot of y = x^2')  # Add title
plt.xlabel('x')  # Label x axis
plt.ylabel('y')  # Label y axis
plt.grid(True)  # Add grid
plt.show()

# Task 2: Generate a bar chart for sales data of four different products
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]
plt.bar(products, sales, color='green')
plt.title('Weekly Sales of Products')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(axis='y')
plt.show()

# Task 3: Create a pie chart for favorite fruits
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
colors = ['red', 'yellow', 'purple', 'brown']
plt.pie(percentages, labels=fruits, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Favorite Fruits')
plt.legend(loc='upper right')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
