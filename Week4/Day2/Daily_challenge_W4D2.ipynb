Rotate an image by 90 degrees

```python
import os
from PIL import Image
path_to_dataset = '/content/flower_images/flower_images'
image_path = os.path.join(path_to_dataset, '0001.png')
original_image = Image.open(image_path)
```

```python
import matplotlib.pyplot as plt

# Display the image
plt.imshow(original_image)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.ndimage import rotate
import numpy as np
# Rotation function
def rotate_image_90_degrees(image):
    # Convert image to numpy array
    image_array = np.array(image)
    # Rotate image by 90 degrees
    rotated_image = rotate(image_array, 90, reshape=False, mode='nearest')
    # Convert rotated image back to PIL Image object
    rotated_image = Image.fromarray(rotated_image)
    return rotated_image

# Rotate the image by 90 degrees
rotated_image = rotate_image_90_degrees(original_image)

# Display the rotated image
plt.imshow(rotated_image)
plt.axis('off')
plt.title('Rotated Image (90 degrees)')
plt.show()
```

```python
# Function to flip an image horizontally and then vertically
def flip_image(image):
    # Convert image to numpy array
    image_array = np.array(image)
    # Flip image horizontally and then vertically
    flipped_image = np.flip(image_array, axis=1)
    flipped_image = np.flip(flipped_image, axis=0)
    # Convert flipped image back to PIL Image object
    flipped_image = Image.fromarray(flipped_image)
    return flipped_image

# Flip the image horizontally and vertically
flipped_image = flip_image(original_image)

# Display the flipped image
plt.imshow(flipped_image)
plt.axis('off')
plt.title('Flipped Image (Horizontal and Vertical)')
plt.show()
```

```python
# Zoom in on the image by scaling it up
zoomed_image = original_image.resize((int(original_image.width * 10.0), int(original_image.height * 5.2)))

# Display the zoomed-in image
plt.imshow(zoomed_image)
plt.axis('off')
plt.title('Zoomed-in Image (Scaled by 10.0x and 5.2x)')
plt.show()
```

```python
def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()
    # Assuming you have the original and augmented images stored in variables original_image and augmented_image
augmented_image = flip_image(original_image)
show_images(original_image, augmented_image, 'Augmented Image')
```
