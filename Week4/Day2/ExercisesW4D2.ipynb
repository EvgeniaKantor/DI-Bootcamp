**Exercise 1**

Type of the data:

1) A company's financial reports stored in an Exel file

Type: Structured data

Explanation: Financial reports typically follow a structured format with defined fields such as revenue, expenses, profit, etc. Storing them in an Excel file maintains this structure.

2) Photographs uploaded to a social media platform.

Type: Unstructured data

Explanation: Photographs contain visual content that lacks a predefined structure. While metadata such as date, location, and camera settings may be attached, the main content (the image itself) is unstructured.

3) A collection of news articles on a website.

Type: Semi-structured data

Explanation: News articles have a semi-structured format with elements like titles, authors, dates, and content. While some parts of the data (e.g., titles and authors) may have a defined structure, the content of the articles themselves can vary widely.

4) Inventory data in a relational database.

Type: Structured data

Explanation: Inventory data in a relational database is structured, with clearly defined fields for items such as product names, quantities, prices, etc.

5) Recorded interviews from a market research study.

Type: Semi-structured data

Explanation: Recorded interviews have a semi-structured format, as they may contain structured metadata such as interviewee names, interview dates, and topics discussed. However, the actual content of the interviews is typically unstructured and can vary from interview to interview.
