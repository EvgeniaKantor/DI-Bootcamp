```python
#Exercise 1 : Linear Regression Analysis
```

```python
import numpy as np
from scipy.stats import linregress
```

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr
```

```python
from scipy import stats
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import binom
```

```python
# Given data
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])
```

```python
# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(house_sizes, house_prices)
```

```python
# Print slope and intercept
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
# Predict the price of a house that is 90 square meters
predicted_price = slope * 90 + intercept
print("Predicted price for a 90 square meter house:", predicted_price)
```

```python
# Create a scatter plot of the data
plt.scatter(house_sizes, house_prices, label='Data')

# Plot the regression line
plt.plot(house_sizes, slope * house_sizes + intercept, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('House Size (sqm)')
plt.ylabel('House Price (currency units)')
plt.title('Linear Regression Analysis')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

The slope of the regression line,
1828.767, represents the rate of change in house price per square meter increase in size. In other words, for each additional square meter increase in the size of a house, the price is expected to increase by approximately
1828.767 currency units, assuming all other factors remain constant.

```python
#Exercise 2 : Understanding ANOVA
# Given data
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
# Perform ANOVA test
f_statistic, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
```

```python
# Print F-value and P-value
print("F-value:", f_statistic)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Based on the P-value, the fertilizers have significantly different effects on plant growth.")
else:
    print("Based on the P-value, there is no significant difference in the effects of the fertilizers on plant growth.")
```

Since the P-value 2.7542672098048448e-05 is much smaller than the significance level (usually
0.05
0.05), we reject the null hypothesis. This means that there is sufficient evidence to conclude that the fertilizers have significantly different effects on plant growth.

```python
# Plotting
plt.figure(figsize=(8, 6))

plt.boxplot([fertilizer_1, fertilizer_2, fertilizer_3], labels=['Fertilizer 1', 'Fertilizer 2', 'Fertilizer 3'])
plt.title('ANOVA: Group Value Distributions')
plt.xlabel('Fertilizer')
plt.ylabel('Plant Growth (cm)')
plt.grid(True)

plt.show()
```

```python
#Exercise 3 : Probability Distributions
# Number of trials (coin flips)
n = 10
# Probability of success (getting a head)
p = 0.5
# Number of successes (getting exactly 5 heads)
k = 5
# Calculate the probability using the binomial distribution
probability = binom.pmf(k, n, p)
# Print the result
print("Probability of getting exactly 5 heads in 10 coin flips:", probability)
```

```python
# Create an array of possible number of successes (from 0 to n)
k_values = np.arange(0, n+1)

# Calculate the probabilities for each number of successes
probabilities = binom.pmf(k_values, n, p)

# Plot the binomial distribution
plt.figure(figsize=(8, 6))
plt.bar(k_values, probabilities, color='skyblue')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
#Exercise 4 : Correlation Coefficients
# Given data
data = pd.DataFrame({'age': [23, 25, 30, 35, 40],
                     'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(data['age'], data['income'])
# Calculate Spearman correlation coefficient
spearman_corr, _ = spearmanr(data['age'], data['income'])

# Print the results
print("Pearson correlation coefficient:", pearson_corr)
print("Spearman correlation coefficient:", spearman_corr)
```

```python
# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['income'], color='blue')
plt.title('Scatter Plot of Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

The Pearson correlation coefficient calculated between the 'age' and 'income' variables is
0.9997, indicating a very strong positive linear relationship between the two variables.

The Spearman correlation coefficient is
1.0, suggesting a perfect monotonic relationship between the variables, where they increase or decrease together perfectly.

Both coefficients confirm a strong association between age and income in the dataset.
